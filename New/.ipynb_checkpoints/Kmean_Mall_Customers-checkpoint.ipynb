```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from KOptimal import KMean
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset.describe()
```

```python
dataset.head()
```

```python
df = dataset.drop(columns=['CustomerID','Gender'])
```

```python
df.head()
```

```python
X = df.values
```

```python
Km = KMean(X)
```

```python
Km.elbow_method(20)
Km.Show_elbow_method()
```

```python
Km.Gap_statistic(20)
Km.Show_Gap_Statistic()
```

```python
Km.silhouette_method(20)
Km.Show_silhoutte_method()
```

```python
Km.fit(7)
```

```python
Km.centroids
```

```python
Labels = Km.labels
```

```python
for i in range(7):
    temp = df.iloc[Labels==i]
    print(temp,sep="\n")
```

```python
df['Labels']=Labels
```

```python
dfl=df.groupby(Labels)
```

```python
dfl.head()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
defi_color=['b','g','r','c','m','y','k','b','g','r','c','m','k','w','b','g','r']
```

```python
%matplotlib qt
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.set_title('Customer')
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Spending Score')
for i in range(7):
    temp = df.iloc[Labels==i]
    ax.scatter(temp['Age'],temp['Annual Income (k$)'],temp['Spending Score (1-100)'],c=defi_color[i])
ax.legend(df.Labels.unique())
plt.show()
```
